# Example: Constructing Training Data

This notebook is intended to be used with the "Constructing Training Data" example in Tecton's documentation. It highlights a few common commands that a user can execute to construct training data.

```
import tecton

# Replace this value with the name you used for your example workspace
workspace_name = 'example_workspace'

# Load the feature package you created in your example workspace
example_workspace = tecton.get_workspace(workspace_name)
example_fs = example_workspace.get_feature_service('ctr_prediction_service')
```

```
events = spark.read.parquet("s3://ad-impressions-data/ctr_events.pq")
display(events)
```

```
training_data = example_fs.get_feature_dataframe(events, timestamp_key="timestamp")
display(training_data.to_spark())
```
