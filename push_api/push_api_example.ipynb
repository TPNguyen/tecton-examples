# Example: Push API

This notebook is intended to be used with the "Pushing Feature Values into the Feature Store" example in Tecton's documentation. It highlights a few common commands that a user can execute to ingest and preview features they have just created.

```
import tecton

fp = tecton.get_feature_package('user_purchases_push_fp')
```

```
import pandas

# Create a single-row DataFrame with example join key and feature value
pandas_df = pandas.DataFrame([{
	"timestamp": pandas.Timestamp("2020-09-18 12:00:06", tz="UTC"),
	"userid": "u123",
	"num_purchases": 92
}])

# Push the Dataframe into the feature package
fp.ingest(pandas_df)
```

```
# Call the preview method to see a sample of the feature data - initially the row we have just pushed
fp.preview()
```

```
# Test online feature fetch
join_keys = {
  "user_id": "u123"
}

print(tecton.get_online_features(fp, join_keys))
```
