# Example: Creating a Feature

This notebook is intended to be used with the "Creating a Feature" example in Tecton's documentation. It highlights a few common commands that a user can execute to preview features they have just created.

```
import tecton

# Replace this value with the name you used for your example workspace
workspace_name = 'example_workspace'

# Load the feature package you created in your example workspace
example_workspace = tecton.get_workspace(workspace_name)
example_fp = example_workspace.get_feature_package('partner_ctr_performance')
```

```
# Call the preview method to see a sample of the feature data
example_fp.preview()
```
